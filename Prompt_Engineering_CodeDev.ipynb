## 1. LLM 이해

### 1-1 LLM이란?

- LLM(Large Language Model)은 자연어 처리를 위해 특별히 설계된 대규모 인공지능 모델입니다. 이 모델들은 대량의 텍스트 데이터에서 패턴을 학습하여, 입력된 문장의 의미를 이해하고, 적절한 대답을 생성하거나 텍스트 요약, 번역, 질문에 답하기 등 다양한 언어 기반 작업을 수행할 수 있습니다.

- LLM은 딥러닝 기술, 특히 트랜스포머라는 구조를 사용합니다. 트랜스포머는 입력된 데이터의 전체를 한 번에 처리하면서 중요한 부분에 주목할 수 있게 해주는 '어텐션 메커니즘'을 활용합니다. 이를 통해 문장이나 문서에서 맥락을 파악하고 더 정확한 응답을 생성할 수 있습니다.

- ChatGPT와 같은 GPT(Generative Pre-trained Transformer)는 먼저 대규모 데이터셋으로 사전 학습을 진행한 후, 특정 작업을 수행하기 위해 추가로 미세 조정을 거칩니다. 이 과정을 통해 특정 주제나 분야에서도 효과적인 대화와 정보 제공이 가능해집니다.

- instruction following : 명령에 따라 행동

```python
'''
양자역학의 불확정성의 원리에 대해 200자 이내로 설명해줘 
'''
```

양자역학의 불확정성의 원리는, 물리학에서 미래의 사건을 정확하게 예측하는 것이 불가능하고, 어떤 시스템의 특정 물리량을 동시에 정확히 알 수 없는 것을 설명합니다. 즉, 어떤 입자의 위치와 운동량, 에너지 등에 대한 정보 중 일부만 정확하게 알 수 있다는 것입니다. 이 원리는 헤이즌버그의 불확정성 원리로도 알려져 있으며, 물리적 현상이 확률적으로 기술되는 양자적 세계에서 중요한 개념입니다. 불확정성의 원리는 우리가 자연세계를 이해하기 위해 양자역학적인 접근이 필요하며, 현대 물리학의 핵심 원리 중 하나입니다.

- in-Context Learning : 문맥(맥락)에 따라 학습

```python
'''
너는 리처드 파인만이고 나는 물리학에 대해 아무것도 모르는 일반인이다.
일반인도 이해하기 쉽게 불활정성 원리에 대해 200자 이내로 사례를 들어서 설명 해 줘라.
'''
```

물리학에서의 불확정성의 원리를 이해하기 쉽게 예시로 설명해 드리겠습니다. 생각해보세요, 우리가 동전을 던진다고 가정해봅시다. 동전이 떨어질 때, 우리는 '앞면' 또는 '뒷면' 중 하나가 나오기를 기대합니다. 하지만 불확정성의 원리에 따라 우리는 정확하게 어떤 면이 나올지 예측할 수 없습니다. 왜냐하면 우리는 던질 때 동전의 초기 상태, 공기 저항 등 많은 요인들을 정확하게 측정할 수 없기 때문입니다. 따라서 우리는 동전이 앞면과 뒷면 중 어느 면이 나올지, 어떤 확률로 결정될지를 예측합니다. 이와 같이 물리적 현상에서 불확정성의 원리는 미래의 결과를 정확하게 예측할 수 없음을 설명하고 있습니다.

```python
'''
여자친구가 평소에 내가 예쁘다고 생각하는 여자 연예인의 사진을 보여주면서 예쁘냐고 물어 봤다.
사실 그 연예인은 여자친구보다 더 예쁘다고 생각한다.
이때 나는 어떻게 대답해야 하는가? 
'''
```

이런 상황에서는 솔직하고 예의를 갖춘 대답을 하는 것이 좋습니다. 상대방의 민감한 감정을 생각하면서 다음과 같이 대답해 볼 수 있습니다:

"너도 아름답고 특별한 사람인데, 나는 그 사진에는 연예인의 매력을 볼 수 있지만 너에게서는 독특하고 사랑스러운 매력을 느껴요. 네가 나에게 가장 예쁘고 소중한 존재야." 

이렇게 해서 상대방의 자존감을 훼손하지 않으면서도 진심을 전할 수 있습니다. 상대방과의 솔직하고 강건한 관계를 유지하는 데 가장 중요한 것은 상대방을 존중하고 어필하는 것입니다.

```python
'''
다음 글을 참고하여 홍길동이 밥을 언제 먹었는지 알려줘라.

"""
새벽녘, 홍길동은 쌀로 만든 따뜻한 국물 요리와 신선한 야채 샐러드로 아침을 시작했다.
향긋한 녹차로 마무리 지으며 하루를 여는 에너지를 얻었다.
정오가 되자, 그는 해산물과 여러 채소가 들어간 김치찌개와 바삭한 김 몇 조각을 즐겼다.
오후 시간에는 과일과 견과류로 달콤하고 건강한 간식을 골랐다. 
저녁이 되어, 홍길동은 구운 닭고기와 다양한 야채를 곁들인 샐러드를 택했다. 
밤이 깇어감에 따라, 그는 가벼운 야식으로 우유와 쿠키를 조금 취했다. 
이렇게 홍길동은 하루 종일 다양한 음식을 취하며 강하고 활기찬 하루를 보냈다.
"""
''' 
```

주어진 글을 토대로 홍길동이 밥을 먹은 시간을 알아보면 다음과 같습니다:

- 아침: 새벽녘에 따뜻한 국물 요리와 야채 샐러드로 아침을 시작하였습니다.
- 정오: 해산물과 채소가 들어간 김치찌개와 김을 즐겼습니다.
- 오후: 과일과 견과류를 선택하여 간식을 즐겼습니다.
- 저녁: 구운 닭고기와 야채를 곁들인 샐러드를 선택했습니다.
- 밤: 우유와 쿠키를 야식으로 조금 취했습니다.

주어진 정보에 따라, 홍길동은 아침, 정오, 오후, 저녁, 밤의 다양한 시간에 음식을 취하였습니다.

- Step-by-Step Reasoning : 단계별 추론

```python
'''
다음 문제를 보고, 답을 말해줘. 단계별로 생각해 봅시다. 

"""
나는 시장에 가서 사과 10개게, 2개는 수리공를 샀습니다.
사과 2개는 이웃에에게 주었습니다.
그리고 가서 사과 5개를 더 사서 1개를 먹었습니다.
나에게는 몇 개의 사과가 남아 있었나요?
"""
'''
```

주어진 문제를 단계별로 생각해보겠습니다:

1. 처음에 시장에 가서 10개의 사과를 샀습니다. 
2. 그 중 2개를 이웃에게 주었으므로, 남은 사과는 10 - 2 = 8개입니다.
3. 또 다른 2개를 수리공에게 주었으므로, 남은 사과는 8 - 2 = 6개입니다.
4. 추가로 5개의 사과를 더 사서, 전체 사과의 개수는 6 + 5 = 11개입니다.
5. 그리고 1개의 사과를 먹었으므로, 남은 사과는 11 - 1 = 10개입니다.

따라서, 최종적으로 나에게는 10개의 사과가 남게 됩니다.

## 2. 코드 개발을 위한 프롬프트 엔지니어링(Prompt Engineering)

### 2-1 프롬프트 엔지니어링 이해

##### 2-1-1 프롬프트 엔지니어링이란?
- 파인튜닝과 프롬프트 엔지니어링은 기계학습 기반의 자연어 처리 모델을 최적화하는 기술입니다. 이러한 기술은 이미 학습된 모델에 새로운 데이터를 반영하여 성능을 향상시키는 데 사용됩니다.
- 이러한 기술은 이미 학습된 모델에 새로운 데이터를 반영하여 성능을 향상시키는 데 사용합니다.
- 파인튜닝은 기존 모델의 가중치를 조정하여 새로운 작업에 맞게 모델을 조정하는 것을 의미하며, 프롬프트 엔지니어링은 모델의 입력에 대한 적절한 프롬프트를 설계하여 모델의 출력을 개선하는 것을 의미합니다.

##### 2-1-2 프롬프트 엔지니어링의 목적
- 파인튜닝은은 대규모의 텍스트 데이터세트를 사용하여 모델을 학습시킨 뒤, 특정 작업에 대해 추가로 학습시키기 때문에 높은 성능을 제공할 수 있으나 학습 시간과 비용, 많은 자원이 증가할 수 있습니다.
- 프롬프트 엔지니어링은 입력을 최적화하여 모델의 성능을 개선하기 때문에 많은 자원,시간,비용 등을 절감하면서 빠르게 모델의 성능을 개선할 수 있음.
- 환각(Hallucination) 현상을 방지할 수 있습니다.

##### 2-1-3 프롬프트의 요소
- Instruction : 모델에게 어떤 작업을 수행할 것인지 알려주는 것입니다.
- Context : 모델에게 어떤 상황에서 작업을 수행할 것인지 알려주는 것입니다.
- Input Data : 모델에게 어떤 데이터를 입력할 것인지 알려주는 것입니다.
- Output Indicator : 모델에게 어떤 결과를 출력할 것인지 알려주는 것입니다.

### 2-3 CoT 프롬프트 엔지니어링

##### 2-3-1 CoT 프롬프트 엔지니어링 이해

- CoT(Chain-of-Thought) 프롬프트 엔지니어링은 자연어 처리 모델, 특히 대규모 언어 모델이 더 복잡하고 추상적인 문제를 해결할 수 있도록 돕는 방법론입니다.
- 이 접근법은 모델에게 문제를 해결하는 과정에서 중간 단계의 사고 과정을 명시적으로 표현하도록 요구함으로써, 모델의 추론 능력을 향상시키는 것을 목표로 합니다.
- 예를 들어, 수학 문제를 해결할 때 CoT 프롬프트 엔지니어링은 모델이 단순히 답만을 제시하는 것이 아니라, 그 문제를 어떻게 해석하고, 어떤 중간 계산을 거쳐 최종 답을 도출했는지 그 과정을 설명하도록 유도합니다.
- 이러한 방식은 모델이 보다 정확하고 신뢰할 수 있는 답변을 제공하도록 돕고, 모델의 추론 과정을 사용자가 이해하기 쉽게 만듭니다.
- 이러한 방식은 특히 교육, 과학적 탐구, 기술적 문제 해결과 같이 고도의 추론이 필요한 분야에서 유용합니다.

##### 2-3-2 CoT 프롬프트 엔지니어링의 주요 장점

- **투명성 향상**: 모델이 어떻게 결론에 도달했는지 그 과정을 명시적으로 보여줌으로써, 모델의 동작이 더 투명해집니다.
- **결정력 강화**: 문제 해결 과정에서 생각하는 단계를 통해, 모델이 더 복잡하고 깊은 문제에 대해 효과적으로 접근할 수 있습니다.
- **사용자 신뢰도 증가**: 모델이 제공하는 답변이 더 정교하고 설득력 있게 다가오므로 사용자의 신뢰도가 증가합니다.

### 3-2 단계별 기능 추가하며 코딩하기

##### 3-2-1 네이버 페이지에서 '반도체'로 검색하여 화면에 출력되는 뉴스 제목 추출하기

- 간략한 프롬프팅 후 코드 수정하는 방법

```python
'''
다음 내용을 파이썬으로 작성하세요.

네이버 홈페이지에 '반도체'라고 검색하여 하단에 출력되는 뉴스의 제목을 추출해줘 
'''
```

```python
# 다음은 파이썬 코드 예시입니다.

import requests
from bs4 import BeautifulSoup

url = "https://www.naver.com"
search_term = "반도체"

# 네이버 홈페이지에 검색어를 입력하여 검색 결과 페이지에 접속
response = requests.get(url + "/search?query=" + search_term)

# HTML에서 뉴스 제목을 추출
soup = BeautifulSoup(response.text, "html.parser")
news_titles = soup.select(".news_tit")

# 추출된 뉴스 제목을 출력
for title in news_titles:
    print(title.get_text())

# 이 코드는 requests 라이브러리를 사용하여 네이버 홈페이지에 접속하고, 검색어를 입력하여 검색 결과 페이지에 접속합니다.
# 그다음 BeautifulSoup 라이브러리를 사용하여 HTML에서 뉴스 제목을 추출하고, 해당 제목을 출력합니다. 코드를 실행하면 네이버에서 '반도체'라는 검색어로 검색한 결과의 뉴스 제목들이 출력됩니다.
```
